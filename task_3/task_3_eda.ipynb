### Applying text cleaning techniques  

-We will perform techniques for cleaning the text such as:
- Normalize text: like case normalization
- Removing unicode Characters : Punctuation, Emoji’s, URL’s and @’s 
- Removing Stopwords : common words within sentences that do not add value 
- Performing Stemming : simpler of the two, groups words by their root stem for e.g (jumping,jumped,jump) = jump
- Performing Lemmatization :groups words based on root definition, and allows us to differentiate between present, past, and indefinite 


```python
# Lets test above techniques with single text data
test_text=train_data['text'][5]
print(test_text)
```

- combine above techniques into one function for ease of use

```python
def clean_text_data(text, stem="None"):

    cleaned_text = ""

    # Make lower
    text = text.lower()

    # Remove line breaks
    text = re.sub(r'\n', '', text)
    
    # Remove Punctuation,URl, and @ 
    
    text_filtered = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", text)
   
    # Remove stop words
    text_filtered = text_filtered.split()
    useless_words = nltk.corpus.stopwords.words("english")

    text_filtered = [word for word in text_filtered if not word in useless_words]
    
    # Remove numbers
    text_filtered = [re.sub(r'\w*\d\w*', '', w) for w in text_filtered]
    # Stem ,Lemmatize
    if stem == 'Stem':
        stemmer = PorterStemmer() 
        text_stemmed = [stemmer.stem(y) for y in text_filtered]
    elif stem == 'Lem':
        lem = WordNetLemmatizer()
        text_stemmed = [lem.lemmatize(y) for y in text_filtered]        
    else:
        text_stemmed = text_filtered
    
    cleaned_text = ' '.join(text_stemmed)

    return cleaned_text
```

```python
# Clean the entire dataset using apply method in pandas
train_data['text']=train_data['text'].apply(clean_text_data,stem = 'Lem')
```

### Word Cloud 

```python
def create_word_cloud(data,type_review):
    data = ' '.join(data)
    plt.figure(figsize = (30,30))
    wc = WordCloud(max_words = 1000, width=1200, height=600,background_color="white").generate(data)
    plt.imshow(wc , interpolation = 'bilinear')
    plt.axis('off')
    plt.title('Word cloud for '+ type_review + ' reviews',fontsize = 40)
    plt.show()

```

```python
# Positive review word  cloud
create_word_cloud(positive_reviews,'positive')
```

## Plotting ngrams

```python
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]
```

```python
top_tri_grams=get_top_ngram(train_data['text'],n=3)
x,y=map(list,zip(*top_tri_grams))
sns.barplot(x=y,y=x)
```
