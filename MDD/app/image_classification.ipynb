Please run the cell below. By executing this cell, the app_docker_nginx image and the app_docker_app image are created and then it will instantiate and run the nginx and deep-learning-production containers respectively.

Once the execution is done, you can send a request along with the name of the image you want do image classification on and get the classification result.

```python
%%bash
docker-compose up
```

The below cell, sends a curl request with the POST method to get classification result from the app. 
Please put your image/s in the data folder of app.
To use your own image, simply replace the last part of the request (i.e. "coco3.png") with your own image name.

```python
%%bash
curl -X POST http://localhost:5000/predict --header 'Content-Type: application/json' -d '{"image name":"coco3.png"}'
```
